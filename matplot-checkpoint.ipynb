# Simple Plot

```python
plt.plot([0,1,2,3],[-1,2,-3,4])
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
#plt.plot(x,y,label='2x',color="Green",linewidth=3,linestyle='--',marker='*',markersize=15,markeredgecolor='Red')
plt.plot(x,y,'r^--')
x2=np.arange(1,3,5)
plt.plot(x2,x2**2)
plt.title("Simple graph",fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([0,2,4,6,8])
plt.yticks([0,2,4,6,8])
plt.legend()
plt.show()# this used to remove [<matplotlib.lines.Line2D at 0x1b74c83fdc0>] see above plot
```

```python
gas=pd.read_csv(r'C:\Users\Gaurav\Desktop\Raj\Matplot\gas_prices.csv')
gas.head()
```

```python
plt.figure(figsize=(10,8))
plt.title('Year v/s Country')
plt.xticks(gas.Year[::3])
plt.xlabel('Years---->',size=15)
plt.ylabel('US dollars---->',size=15)
#print(gas.Year[::3])
plt.plot(gas.Year,gas.Canada,'ro-',label='CANADA',color='Green',markeredgecolor='Red')
plt.plot(gas.Year,gas.USA,'ro-',label='USA',color='Orange',markeredgecolor='Black')
plt.plot(gas.Year,gas['South Korea'],'ro-',label='South Korea',color='#00ff00',markeredgecolor='Black')
plt.plot(gas.Year,gas.Australia,'ro-',label='Australia',color='Blue',markeredgecolor='Red')
plt.savefig('Gas price.png',dpi=300)
plt.legend()
plt.show()
```

```python
for country in gas:
    if country!='Year':
        plt.xticks(gas.Year[::3])
        plt.plot(gas.Year,gas[country],marker='*')
plt.legend()
plt.show()
        
```

```python
coun_fil=['Australia','Japan','South Korea']
for columns in gas:
    if columns in coun_fil:
        plt.plot(gas.Year,gas[columns])
plt.legend()
plt.show()
    
```
